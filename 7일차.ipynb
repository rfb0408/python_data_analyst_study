```python
# 머신러닝에서 문제집과 정답지는 병렬 계산이 가능해야한다.
# numpy 쓰는 이유: 주어진 값이 리스트 형식이여서 병렬 계산이 가능한 형태인 행렬로 바꾸기 위해서 사용한다. 이때 array함수가 쓰인다.
# 예) [2,4,6,8]*2를 시행해서 [4,8,12,16]을 얻고 싶은데, [2,4,6,8,2,4,6,8]이 나온다. 원하는 값을 얻게 해주는 것이 numpy
# 공부 시간과 시험 성적 데이터를 가지고 분석해보자!

import numpy as np
import matplotlib.pyplot as plt

x1=np.array([2,4,6,8])
y=np.array([20,70,55,80])

a1=0 # 기울기
b=0 # 절편
```

```python
plt.scatter(x1,y)
plt.plot(x1,a1*x1+b,"red")
```

```python
import numpy as np

x1=np.array([2,4,6,8])
y=np.array([20,70,55,80])

a1=0
b=0

lr=0.005 #학습률! 커도 완되고, 너무 작아서도 안된다

for n in range (3000) :

    y_hat=a1*x1+b
    error = y_hat-y

    a1_diff=sum(2*error*x1)
    b_diff=sum(2*error*1)

    a1=a1-a1_diff *lr
    b = b-b_diff*lr

    print(round(a1_diff,2),round(b_diff,2)) #round 반올림 함수
```

```python
a1*5+b # 구해낸 데이터로 5시간동안 공부한 학생의 성적을 유추할 수 있다.
```

```python
# 변수를 추가해 보자
x1=np.array([2,4,6,8])
x2=np.array([1,2,1,3]) # 문제집 푼 갯수
y=np.array([20,70,55,80])

a1=0
a2=0
b=0
lr=0.005


for n in range (3000) :

    y_hat=a1*x1+a2*x2+b #요인 추가되면 이렇게 바뀐다.
    error = y_hat-y

    a1_diff=sum(2*error*x1)
    a2_diff=sum(2*error*x2)

    b_diff=sum(2*error*1)

    a1=a1-a1_diff *lr
    a2=a2-a2_diff *lr

    b = b-b_diff*lr

    print(round(a1_diff,2),round(b_diff,2),round(a2_diff,2))
```

```python
plt.scatter(x1,y)
plt.plot(x1,a1*x1+a2*x2+b,"red")
```

```python
a1*3+a2*2+b # 3시간 동안 2개의 문제집을 푼 학생의 점수 예측
```

## sin 함수 머신러닝

```python
# sin 함수 그리기
import numpy as np

x=np.arange(-3,3,0.01) # -3부터 3까지 0.01씩  증가시켜라
y=np.sin(x)

a=0
b=0
```

```python
plt.plot(x,y)
plt.plot(x,a*x+b,"red")
```

```python
# sin 함수와 오차가 가장 작은 직선 그리기
import numpy as np
import matplotlib.pyplot as plt

a=0
b=0
lr=0.0001

for n in range (5000) :

    y_hat=a*x+b
    error = y_hat-y

    a_diff=sum(2*error*x)
    b_diff=sum(2*error*1)

    a=a-a_diff *lr
    b=b-b_diff *lr

    print(round(a_diff,2),round(b_diff,2))
```

```python
plt.scatter(x,y)
plt.plot(x,a*x+b,"red")
```

```python
# sin 함수와 오차가 가장 2차 함수 곡선 그리기

a=0
b=0
c=0
lr=0.00001

for n in range (5000) :

    y_hat=a*x**2+ b*x + c
    error = y_hat-y

    a_diff=sum(2*error*x**2)
    b_diff=sum(2*error*x)
    c_diff=sum(2*error*1)

    a=a-a_diff *lr
    b=b-b_diff *lr
    c=c-c_diff *lr

    print(round(a_diff,2),round(b_diff,2),round(c_diff,2))
```

```python
plt.scatter(x,y)
plt.plot(x,a*x**2+b*x+c,"red")
```

```python
# 이차함수 곡선을 그렸는데 직선이 나오는 이유 : a와 c의 값이 너무 작아서 값을 날려버림
a,b,c
```

```python
# sin 함수와 오차가 가장 3차 함수 곡선 그리기

a=0
b=0
c=0
d=0
lr=0.00001

for n in range (5000) :

    y_hat=a*x**3+ b*x**2+c*x+d
    error = y_hat-y

    a_diff=sum(2*error*x**3)
    b_diff=sum(2*error*x**2)
    c_diff=sum(2*error*x)
    d_diff=sum(2*error*1)

    a=a-a_diff *lr
    b=b-b_diff *lr
    c=c-c_diff *lr
    d=d-d_diff *lr

    print(round(a_diff,2),round(b_diff,2),round(c_diff,2),round(d_diff,2))
```

```python
plt.scatter(x,y)
plt.plot(x,a*x**3+b*x**2+c*x+d,"red")
```

### 회귀모델과 분류모델

K 최근접 이웃 회귀 모델 : KNeighborsRegressor

```python
# 회귀모델 : 연속적인 r값
# K 최근접 이웃 회귀 모델
from sklearn.neighbors import KNeighborsRegressor

perch = pd.read_csv("perch.csv")

data = perch.iloc[:,:-1].to_numpy()
target = perch['weight'].to_numpy()

train_x,test_x,train_y,test_y = train_test_split(data,target,test_size=0.2)

mean = np.mean(train_x,axis=0)
std=np.std(train_x,axis=0)

train_scaled = (train_x-mean) /std
test_scaled = (test_x-mean) /std

model = KNeighborsRegressor() # 가장 가까운 5개의 데이터를 찾아옴
model.fit(train_scaled,train_y)

box=[]
for n in range(1,21): # 찾는 이웃을 달리해서 정확도를 높여보

    model.n_neighbors=n # 모델이 기본적으로는 5개지만 n개의 이웃을 찾도록 해준다.
    score=model.score(test_scaled,test_y)
    box.append(score)

model=KNeighborsRegressor(n_neighbors=4) # 찾는 이웃을 4개로 바꿔준 다음에는 훈련 데이터와 시험데이터를 전부 학습시켜도 된다
model.fit(train_scaled,train_y)
```

```python
plt.plot(range(1,21),box)
```

```python
model.score(test_scaled,test_y)
```
