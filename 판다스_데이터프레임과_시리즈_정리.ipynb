#### 3-2. 행/열 접근하기
* 레이블을 사용한 행 선택 (loc)  
* 인덱스를 사용한 행 선택 (iloc)  

```python
# loc : 이름으로 접근
# iloc : 순서로 접근

df.loc[0] # 0이라고 적힌 행에 접근
```

```python
df.loc[[0,10,100,1000]] # 0,10,100,1000이라고 적힌 행에 접근
```

```python
df.loc[[0,10,100,1000],['country','year']]
```

```python
df.iloc[0] #0번째 행에 접근
```

```python
df.iloc[[0,10,100,1000]] # 0,10,100,1000번째 행에 접근
```

```python
df.iloc[[0,10,100,1000],[0,2]]
```

```python
df.loc[4:8] # 얘는 8행까지 가져옴!!!
```

```python
df.iloc[4:8] # 얘는 7행까지 가져옴 !!
```

#### 3-3. 특정 조건에 따른 행 접근하기
* 특정 조건을 만족하는 행만 선택할 수도 있습니다.
* query 메소드를 이용한 복잡한 조건 선택할 수도 있습니다.

```python
# 평균, 중앙값이 데이터를 분석하는데 있어서 중요하다. 그러므로 누락값과 이상치가 있는지를 판단해야한다!
# info함수,describe
df.info() # 지금 내 데이터가 어떠한지 확인
```

```python
df.describe() # 갯수, 평균, 최대, 최소 다 나옴
```

#### 3-4. 특정 값이 포함된 행 접근하기
* isin 메소드를 사용하여 열의 값이 리스트에 있는 값들 중 하나와 일치하는 행을 선택합니다.

```python
# 기대수명의 값이 50 이상인 것만 접근

A = df["lifeExp"]>50
A
```

```python
df[A]
```

```python
A = df["lifeExp"]>50
B=df['continent'] == 'Europe'
C=df['continent'] == 'Asia'
D=df['continent'] == 'Africa'

df[A&B] # & :교집합, |:합집합, (\를 shift눌러서 하면 | 가 나온다)
```

```python
df[B|C|D]
```

```python
# 기대수명이 기대수명의 평균보다 큰 행만 접근하기
df[df['lifeExp']>df['lifeExp'].mean()]
```

```python
df[df['continent'].isin(['Europe','Asia','Africa'])] #열의 값이 리스트에 있는 값들 중 하나와 일치하는 행을 선택!
```

```python
df[~df['continent'].isin(['Europe','Asia','Africa'])] #~반대
```

#### 3-5. 데이터프레임 값 필터링하기
* where 메소드를 사용하여 조건을 기반으로 데이터프레임의 값을 필터링합니다.

```python
df.where(df["lifeExp"]>60) # 조건문을 안에 적는 것과 다른점은 NaN 값을 남긴다는 점이다
```

```python
df["lifeExp"].where(df["lifeExp"]>60,10) # 60보다 작은 행은 10을 채워넣은다
```

#### 3-6. str 접근자
* str 메서드는 시리즈(Series) 또는 데이터프레임(DataFrame)의 각 요소에 대해 문자열 연산을 수행합니다.

```python
# .str //시리즈 안에 있는 값으로 접근한다.
df['country'].str.upper() #'country' 안에 접근해서 전부 대문자로 바꿔
```

```python
df['country'].str.startswith('Z') #문자열이 특정 문자 또는 문자열로 시작하거나 끝나는지 검사
```

```python
# 데이터베이스 : sql

df.query("lifeExp>60&continent =='Africa'")
```
